Para llevar a cabo este proyecto contamos con una base de datos con información de los extranjeros que han ingresado al país (Colombia) desde el 2012 hasta la actualidad, la cual nos permitirá analizar la cantidad de extranjeros que ingresaron al país en un periodo de tiempo determinado, su nacionalidad y que nacionalidades son más comunes, todo esto para poder crear estrategias turisticas y plantear hipótesis acerca del motivo de su visita.

**Pregunta de investigación:** ¿Cuáles son los patrones y tendencias de la entrada de extranjeros a Colombia en términos de género, nacionalidad y ubicación geográfica?

**Posibles Objetivos:**
1. Analizar la distribución anual de la entrada de extranjeros a Colombia.
2. Investigar las diferencias en la entrada de extranjeros según su nacionalidad y determinar los países de origen más comunes.
3. Examinar la proporción de hombres, mujeres e individuos de género indefinido entre los extranjeros que ingresan a Colombia.
4. Explorar la relación entre la ubicación geográfica y la entrada de extranjeros, identificando las regiones o ciudades con mayor afluencia migratoria.
5. Identificar patrones teniendo en cuenta el mes de ingreso al país.

```python
Extranjeros_Col['Año'].value_counts().sort_index()
frecuencia_por_año = Extranjeros_Col['Año'].value_counts().sort_index()
print("Distribución anual de entrada de extranjeros:")
print(Extranjeros_Col['Año'].value_counts().sort_index())
frecuencia_por_año.plot(kind='line', marker='o', color='purple')
```

```python
nacionalidades_comunes = Extranjeros_Col['Nacionalidad'].value_counts().head()
print("\nPaíses de origen más comunes:")
print(nacionalidades_comunes)
nacionalidades_comunes.plot(kind='bar', color='pink')
```

```python
genero_counts = Extranjeros_Col[['Femenino', 'Masculino', 'Indefinido']].sum()
print("\nProporción de género entre los extranjeros:")
print(genero_counts)
genero_counts.plot(kind='bar', color=['yellow', 'teal', 'gray'])
```

```python
ciudades_con_mas_ingresos = Extranjeros_Col['Ciudad'].value_counts().head()
print("\nCiudades con mayor afluencia migratoria:")
print(ciudades_con_mas_ingresos)
ciudades_con_mas_ingresos.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporción de Ingresos por Ciudad')
plt.show()
```

```python
ciudades_con_menos_ingresos = Extranjeros_Col.groupby('Ciudad')['Total'].sum().sort_values().head()
print("\nCiudades con menor afluencia migratoria:")
print(ciudades_con_menos_ingresos)
ciudades_con_menos_ingresos.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporción de Ingresos por Ciudad')
plt.show()
```

```python
ingresos_por_mes = Extranjeros_Col.groupby('Mes').size()
print("\nPatrones de ingreso por mes:")
print(ingresos_por_mes)
colores= ['skyblue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta', 'yellow', 'olive']
ingresos_por_mes.plot(kind='bar', color=colores)
```

```python
paises = ["Venezuela", "España", "Estados Unidos de américa","Reino Unido de Gran Bretaña e Irlanda del Norte","Países Bajos"]
meses = ["Enero", "Febrero", "Marzo","Abril","Mayo","Junio","Julio","Agosto","Septiembre","Octubre","Noviembre","Diciembre"]
filtro_multiple = Extranjeros_Col[
    (Extranjeros_Col["Nacionalidad"].isin(paises)) &
    (Extranjeros_Col["Mes"].isin(meses)) &
    (Extranjeros_Col["Año"] == 2023)
]
print(filtro_multiple)
resultados_agrupados = filtro_multiple.groupby(["Nacionalidad", "Mes"])["Total"].sum()
print(resultados_agrupados)
resultados_agrupados.unstack().plot(kind='bar', width=0.8)
```

```python
Extranjeros_Col[['Femenino', 'Masculino']] = Extranjeros_Col[['Femenino', 'Masculino']].astype(int)
afluencia_femenino_ciudad = Extranjeros_Col[Extranjeros_Col['Femenino'] > 0].groupby('Ciudad')[['Femenino']].sum().reset_index()
afluencia_femenino_ciudad['Total'] = afluencia_femenino_ciudad['Femenino']
afluencia_femenino_ciudad = afluencia_femenino_ciudad.sort_values(by='Total', ascending=False)
afluencia_femenino_ciudad = afluencia_femenino_ciudad.head(5)
print(afluencia_femenino_ciudad)
ciudades_femenino = afluencia_femenino_ciudad['Ciudad']
total_femenino = afluencia_femenino_ciudad['Total']
plt.figure(figsize=(10, 6))
plt.barh(ciudades_femenino, total_femenino, color='yellow')
plt.xlabel('Total de Afluencia Femenina')
plt.title('Afluencia de Extranjeras por Ciudad (Top 5)')
plt.gca().invert_yaxis()
plt.show()
```

```python
Extranjeros_Col[['Femenino', 'Masculino']] = Extranjeros_Col[['Femenino', 'Masculino']].astype(int)
afluencia_masculino_ciudad = Extranjeros_Col[Extranjeros_Col['Masculino'] > 0].groupby('Ciudad')[['Masculino']].sum().reset_index()
afluencia_masculino_ciudad['Total'] = afluencia_masculino_ciudad['Masculino'] + afluencia_masculino_ciudad['Masculino']
afluencia_masculino_ciudad = afluencia_masculino_ciudad.sort_values(by='Total', ascending=False)
afluencia_masculino_ciudad = afluencia_masculino_ciudad.head(5)
print(afluencia_masculino_ciudad)
plt.figure(figsize=(10, 6))
plt.barh(afluencia_masculino_ciudad['Ciudad'], afluencia_masculino_ciudad['Total'], color='Teal')
plt.xlabel('Total de Afluencia Masculina')
plt.title('Afluencia de Extranjeros por Ciudad (Top 5)')
plt.gca().invert_yaxis()
plt.show()
```
